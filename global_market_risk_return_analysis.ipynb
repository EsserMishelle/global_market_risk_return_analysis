The dataset represented multiple financial instrument categories, including equities, ETFs, cryptocurrencies, forex, and commodities. Therefore, asset class was selected as the primary categorical feature for comparative analysis rather than sector classifications, which contained substantial missing values.

```python
corr_features = [
    'return_1w_pct',
    'return_1m_pct',
    'return_3m_pct',
    'return_6m_pct',
    'return_1y_pct',
    'volatility_30d_ann',
    'volatility_90d_ann',
    'volatility_1y_ann',
    'sharpe_1y',
    'market_cap_usd'
]
```

```python
corr_matrix = df[corr_features].corr()
```

```python
corr_matrix.rename(index={
    'return_1w_pct':'Return 1W',
    'return_1m_pct':'Return 1M',
    'return_3m_pct':'Return 3M',
    'return_6m_pct':'Return 6M',
    'return_1y_pct':'Return 1Y',
    'volatility_30d_ann':'Volatility 30D',
    'volatility_90d_ann':'Volatility 90D',
    'volatility_1y_ann':'Volatility 1Y',
    'sharpe_1y':'Sharpe 1Y',
    'market_cap_usd':'Market Cap USD'
}, columns={
    'return_1w_pct':'Return 1W',
    'return_1m_pct':'Return 1M',
    'return_3m_pct':'Return 3M',
    'return_6m_pct':'Return 6M',
    'return_1y_pct':'Return 1Y',
    'volatility_30d_ann':'Volatility 30D',
    'volatility_90d_ann':'Volatility 90D',
    'volatility_1y_ann':'Volatility 1Y',
    'sharpe_1y':'Sharpe 1Y',
    'market_cap_usd':'Market Cap USD'
}, inplace=True)
```

```python
plt.figure(figsize=(10, 4))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='Paired',
    fmt='.2f',
    alpha=0.7,
    linewidths=.5,
    annot_kws={"size": 8},
    # linecolor='white',
    cbar_kws={"pad": 0.01}
    # cbar=False
)
plt.xticks(rotation=30)
plt.yticks(rotation=0)
plt.title('Gloabal Metric Correlation Heatmap')
plt.tight_layout()

file_path = '/content/drive/MyDrive/das-consultants/return_vs_risk_heatmap.jpg'
plt.savefig(file_path, dpi=300, bbox_inches='tight')

plt.show()
```
